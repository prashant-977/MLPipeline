```python
dbutils.library.restartPython()
storage_account_key = "<SA>"
container_name = "raw"
storage_account_name = "sadlallstate"
mount_point = "/mnt/datalake"

# Unmount if already mounted
try:
    dbutils.fs.unmount(mount_point)
    print(f"Unmounted {mount_point}")
except:
    print(f"{mount_point} was not mounted")

# Mount
try:
    dbutils.fs.mount(
        source = f"wasbs://{container_name}@{storage_account_name}.blob.core.windows.net",
        mount_point = mount_point,
        extra_configs = {f"fs.azure.account.key.{storage_account_name}.blob.core.windows.net": storage_account_key}
    )
    print("Mount successful!")
except Exception as e:
    print(f"Mount failed: {e}")

# Verify - different approach
try:
    files = dbutils.fs.ls(mount_point)
    if len(files) == 0:
        print("✓ Mount successful but container is empty")
    else:
        print(f"✓ Found {len(files)} items:")
        for file in files:
            print(f"  - {file.name}")
except Exception as e:
    print(f"✗ Error accessing mount: {e}")
```
